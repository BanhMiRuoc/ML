```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets
```

```python
X_xor = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y_xor = np.array([0, 1, 1, 0])
```

```python
model_xor = svm.SVC(kernel='rbf', gamma='auto', C=1)
model_xor.fit(X_xor, y_xor)

xx, yy = np.meshgrid(np.linspace(-0.5, 1.5, 100), np.linspace(-0.5, 1.5, 100))
Z = model_xor.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
```

```python
plt.scatter(X_xor[y_xor == 0][:, 0], X_xor[y_xor == 0][:, 1], color='red', marker='s', label="Class 0")
plt.scatter(X_xor[y_xor == 1][:, 0], X_xor[y_xor == 1][:, 1], color='blue', marker='o', label="Class 1")

plt.contourf(xx, yy, Z, levels=np.linspace(Z.min(), 0, 7), cmap=plt.cm.PuOr, alpha=0.5)
plt.contour(xx, yy, Z, levels=[0], linewidths=2, colors='k')

plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.xlabel('$x_1$', fontsize=14)
plt.ylabel('$x_2$', fontsize=14)
plt.legend()
plt.title("Classifier for XOR")
plt.show()
```
